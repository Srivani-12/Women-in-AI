```python
import pandas as pd
df_covid=pd.read_excel("/content/Covid_Data.xlsx")
```

```python
df_covid
```

```python
df_covid.info()
```

```python
df_covid.State.value_counts().plot(kind="bar")
```

```python
import seaborn as sns
covid_state=df_covid.groupby(["State"],as_index=False)["Cured"].sum().sort_values(ascending=False,by="Cured").head()
covid_state
sns.barplot(x="State",y="Cured",data=covid_state)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
covid_active=df_covid.groupby(["State"],as_index=False)["Active Cases"].sum().sort_values(ascending=False,by="Active Cases").head()
covid_active
sns.barplot(x="State",y="Active Cases",data=covid_active)
plt.xlabel("States")
plt.ylabel("Active Cases")
plt.title("States vs active cases")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df_new=df_covid.groupby('State')["Active Cases"].sum()
df_new=df_new.sort_values(ascending=False).head(5)
df_new.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df_new=df_covid.groupby('State')["Active Cases"].sum()
df_new=df_new.sort_values(ascending=False).head(10)
df_new.plot(kind="bar")
plt.xlabel("Active Cases")
plt.ylabel("State")
plt.title("Active cases vs State")
plt.show()
```

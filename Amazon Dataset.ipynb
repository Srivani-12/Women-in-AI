```python
#write a function that will give sum of numbers in the list without using any predefined python function
def list_sum(x):
  sum=0
  for i in x:
    sum=sum+i
  return sum
print(list_sum([1,2,3,4,5]))
```

Lambda Function

```python
#write the function that will give count of vowels present in the given string/
#not to use predefined fuctions/librabries
def count_vowels(x):
  c=0
  y=['a','e','i','o','u']
  for i in x:
    c=lambda c:c+1 if i in y else 0
  return c
count_vowels("bdiugfqjbw")
```
